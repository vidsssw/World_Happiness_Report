# Creating Linear Regression **Model**

```r
numerical_data <- df[c("Happiness.Score","Freedom","Dystopia.Residual","Trust..Government.Corruption.","Generosity","Economy..GDP.per.Capita.","Health..Life.Expectancy.")]
```

```r
set.seed(123)
training.samples <- numerical_data$Happiness.Score %>%
  createDataPartition(p = 0.8, list = FALSE)
train.data  <- numerical_data[training.samples, ]
test.data <- numerical_data[-training.samples, ]
```

```r
model <- lm(Happiness.Score~Health..Life.Expectancy.+Freedom+Dystopia.Residual+Trust..Government.Corruption.+Generosity+Economy..GDP.per.Capita.+Health..Life.Expectancy., data = train.data)
summary(model)
```

```r
coef(model)
```

```r
plot(model)
```

# Evaluation

```r
# Make predictions
predictions <- model %>% predict(test.data)
# Model performance
# (a) Compute the prediction error, RMSE
print(predictions)
```

Mean Absolute Error

```r
MAE(predictions, test.data$Happiness.Score)
```

Root Mean Square Error

```r
RMSE(predictions, test.data$Happiness.Score)
```

Adjusted R2 Score

```r
R2(predictions, test.data$Happiness.Score)
```
